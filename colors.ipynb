## Tâches dans ce notebook
- Affecter la bonne valeur (chiffre) à chaque lettre pour que
    - ISA + ROA = TELO

```python
from itertools import permutations

def valid_solution(lval):
    isa = lval["I"] * lval["S"] * lval["A"]
    roa = lval["R"] * lval["O"] * lval["A"]
    telo = lval["T"] * lval["E"] * lval["L"] * lval["O"]
    return isa, roa, telo, isa + roa == telo


def find_sol(verbose=False, rng=range(10)):
    solutions = []
    for perm in permutations(rng, 8):
        sol_t = {c: i for c, i in zip("ISAROTEL", perm)}
        isa, roa, telo, is_sol = valid_solution(sol_t)
        if is_sol:
            if verbose:
                print(
                    "solution: ",
                    f"{"*".join([str(sol_t[c])for c in "ISA"])} +",
                    f"{"*".join([str(sol_t[c])for c in "ROA"])} ==",
                    f"{"*".join([str(sol_t[c])for c in "TELO"])}",
                    f"=> {isa}+{roa}={telo}"
                )
            else:
                solutions.append(sol_t)
                
    return solutions
```

```python
NUM_RNG = range(1,10)

SOLUTIONS = find_sol(rng=NUM_RNG)
assert SOLUTIONS != [], "Did  not find any solution !!"

print(f"Found {len(SOLUTIONS)} solutions !!")
```
